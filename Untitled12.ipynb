```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Customers = pd.read_excel("Sales_Dataset.xlsx",sheet_name="Customers")
Orders = pd.read_excel("Sales_Dataset.xlsx",sheet_name="Orders")
OrderDetails = pd.read_excel("Sales_Dataset.xlsx",sheet_name="OrderDetails")
Products = pd.read_excel("Sales_Dataset.xlsx",sheet_name="Products")
```

### Calculate total sales per customer.

```python
data = OrderDetails.merge(Products,on="ProductID")
data = data.merge(Orders,on="OrderID")
data = data.merge(Customers,on="CustomerID")
data["Sales"] = data["Price"] * data["Quantity"]
Top_Customers = data.groupby("CustomerName")["Sales"].sum().reset_index()
print(Top_Customers.head(10))
```

```python
plt.figure(figsize=(10,6))
plt.bar(Top_Customers["CustomerName"], Top_Customers["Sales"], color="skyblue")
plt.xticks(rotation=45)
plt.title("Total Sales Top 10 Customers")
plt.ylabel("Sales ($)")
plt.xlabel("Customer Name")
plt.show()
```

### Draw a chart showing monthly sales.


```python
data["OrderDate"] = pd.to_datetime(data["OrderDate"])
data["YearMonth"] = data["OrderDate"].dt.to_period("M")
Monthly_Sales = data.groupby("YearMonth")["Sales"].sum().reset_index()
print(Monthly_Sales)
```

```python
plt.figure(figsize=(10,5))
plt.plot(Monthly_Sales["YearMonth"].astype(str), Monthly_Sales["Sales"], marker="o")
plt.xticks(rotation=45)
plt.title("Monthly_Sales")
plt.ylabel("Sales $" )
plt.xlabel("Month")
plt.show()
```

### Showcase your top 3 highest-grossing products.

```python
Top_Products = data.groupby("ProductName")["Sales"].sum().reset_index().sort_values(by="Sales",ascending=False).reset_index(drop=True)
print(Top_Products.head(3))
```

```python
plt.figure(figsize=(10,6))
plt.bar(Top_Products["ProductName"],Top_Products["Sales"],color="orange")
plt.title("Top 3 Products by Revenue")
plt.ylabel("Sales ($)")
plt.xlabel("Product")
plt.tight_layout()
plt.show()
```
